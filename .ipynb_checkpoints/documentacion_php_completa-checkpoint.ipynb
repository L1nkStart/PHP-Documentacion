# üìò Documentaci√≥n Completa de PHP

Esta documentaci√≥n interactiva cubre todos los aspectos b√°sicos de PHP con ejemplos ejecutables.

**Versi√≥n de PHP utilizada:** 8.2.12  
**Entorno:** Windows con XAMPP  
**Ejecuci√≥n:** Python + subprocess para ejecutar PHP

---

```python
import subprocess
import tempfile
import os

def ejecutar_php(codigo_php):
    """Ejecuta c√≥digo PHP y devuelve la salida"""
    with tempfile.NamedTemporaryFile(mode='w', suffix='.php', delete=False) as f:
        f.write(codigo_php)
        archivo_temp = f.name
    
    try:
        resultado = subprocess.run(
            ['php', archivo_temp],
            capture_output=True,
            text=True,
            encoding='utf-8'
        )
        
        if resultado.returncode == 0:
            return resultado.stdout
        else:
            return f"Error: {resultado.stderr}"
    finally:
        os.unlink(archivo_temp)

try:
    version = subprocess.run(['php', '-v'], capture_output=True, text=True)
    print("‚úÖ PHP disponible:")
    print(version.stdout.split('\n')[0])
except FileNotFoundError:
    print("‚ùå PHP no est√° disponible en el PATH")
```

## 1. üöÄ Introducci√≥n a PHP

PHP (PHP: Hypertext Preprocessor) es un lenguaje de programaci√≥n de prop√≥sito general especialmente adecuado para el desarrollo web.

### Caracter√≠sticas principales:
- **Tipado din√°mico**: Las variables no necesitan declaraci√≥n de tipo
- **Interpretado**: No requiere compilaci√≥n
- **Multiplataforma**: Funciona en Windows, Linux, macOS
- **Gran ecosistema**: Miles de librer√≠as y frameworks

```python
codigo = '''
<?php
echo "¬°Hola, mundo desde PHP!\n";
echo "Versi√≥n de PHP: " . phpversion() . "\n";
echo "Fecha y hora actual: " . date('Y-m-d H:i:s') . "\n";
echo "Sistema operativo: " . PHP_OS . "\n";
?>
'''

resultado = ejecutar_php(codigo)
print("üì§ Salida:")
print(resultado)
```

## 2. üì¶ Variables y Tipos de Datos

En PHP, las variables comienzan con `$` y son de tipado din√°mico.

### Tipos principales:
- **string**: Cadenas de texto
- **int**: N√∫meros enteros
- **float**: N√∫meros decimales
- **bool**: Verdadero/falso
- **array**: Arrays
- **object**: Objetos
- **null**: Valor nulo

```python
codigo = '''
<?php
$nombre = "Mar√≠a Garc√≠a";
$edad = 28;
$altura = 1.68;
$esEstudiante = true;
$direccion = null;

echo "=== INFORMACI√ìN PERSONAL ===\n";
echo "Nombre: $nombre\n";
echo "Edad: $edad a√±os\n";
echo "Altura: {$altura}m\n";
echo "Es estudiante: " . ($esEstudiante ? "S√≠" : "No") . "\n";
echo "Direcci√≥n: " . ($direccion ?? "No especificada") . "\n";

echo "\n=== TIPOS DE VARIABLES ===\n";
echo "Tipo de nombre: " . gettype($nombre) . "\n";
echo "Tipo de edad: " . gettype($edad) . "\n";
echo "Tipo de altura: " . gettype($altura) . "\n";
echo "Tipo de esEstudiante: " . gettype($esEstudiante) . "\n";
echo "Tipo de direccion: " . gettype($direccion) . "\n";
?>
'''

resultado = ejecutar_php(codigo)
print("üì§ Salida:")
print(resultado)
```

```python
codigo = '''
<?php
$texto1 = "Hola";
$texto2 = "Mundo";

echo "=== OPERACIONES CON STRINGS ===\n";
echo "Concatenaci√≥n: " . $texto1 . " " . $texto2 . "\n";
echo "Longitud: " . strlen($texto1) . " caracteres\n";
echo "May√∫sculas: " . strtoupper($texto1) . "\n";
echo "Min√∫sculas: " . strtolower($texto2) . "\n";
echo "Primera letra may√∫scula: " . ucfirst($texto1) . "\n";

$frase = "PHP es un lenguaje genial";
echo "\nFrase original: $frase\n";
echo "Palabras: " . str_word_count($frase) . "\n";
echo "Reemplazar: " . str_replace("genial", "fant√°stico", $frase) . "\n";
echo "Subcadena: " . substr($frase, 0, 3) . "\n";
?>
'''

resultado = ejecutar_php(codigo)
print("üì§ Salida:")
print(resultado)
```

## 3. üóÇÔ∏è Arrays (Arreglos)

PHP tiene arrays muy potentes que pueden ser indexados o asociativos.

### Tipos de arrays:
- **Indexados**: Usan n√∫meros como √≠ndices
- **Asociativos**: Usan strings como claves
- **Multidimensionales**: Arrays dentro de arrays

```python
codigo = '''
<?php
$frutas = ["manzana", "banana", "naranja", "uva", "kiwi"];
$numeros = [10, 25, 33, 47, 52];

echo "=== ARRAYS INDEXADOS ===\n";
echo "Frutas:\n";
foreach ($frutas as $index => $fruta) {
    echo "[$index] => $fruta\n";
}

echo "\nN√∫meros:\n";
foreach ($numeros as $numero) {
    echo "$numero ";
}
echo "\n\nTotal de frutas: " . count($frutas) . "\n";
echo "Primera fruta: " . $frutas[0] . "\n";
echo "√öltima fruta: " . end($frutas) . "\n";
?>
'''

resultado = ejecutar_php(codigo)
print("üì§ Salida:")
print(resultado)
```

```python
codigo = '''
<?php
$persona = [
    "nombre" => "Carlos Ruiz",
    "edad" => 32,
    "ciudad" => "Madrid",
    "profesion" => "Desarrollador",
    "salario" => 45000
];

echo "=== ARRAY ASOCIATIVO ===\n";
foreach ($persona as $clave => $valor) {
    echo "$clave: $valor\n";
}

echo "\n=== ACCESO ESPEC√çFICO ===\n";
echo "El nombre es: " . $persona["nombre"] . "\n";
echo "Vive en: " . $persona["ciudad"] . "\n";

$persona["email"] = "carlos@example.com";
echo "\nDespu√©s de agregar email:\n";
echo "Email: " . $persona["email"] . "\n";
echo "Total de campos: " . count($persona) . "\n";
?>
'''

resultado = ejecutar_php(codigo)
print("üì§ Salida:")
print(resultado)
```

```python
codigo = '''
<?php
$estudiantes = [
    [
        "nombre" => "Ana L√≥pez",
        "edad" => 20,
        "notas" => [8.5, 9.0, 7.5, 8.8]
    ],
    [
        "nombre" => "Luis Garc√≠a",
        "edad" => 22,
        "notas" => [7.0, 8.5, 9.2, 8.0]
    ],
    [
        "nombre" => "Sof√≠a Mart√≠n",
        "edad" => 21,
        "notas" => [9.5, 9.8, 8.7, 9.3]
    ]
];

echo "=== ARRAY MULTIDIMENSIONAL ===\n";
foreach ($estudiantes as $index => $estudiante) {
    echo "Estudiante " . ($index + 1) . ":\n";
    echo "  Nombre: " . $estudiante["nombre"] . "\n";
    echo "  Edad: " . $estudiante["edad"] . " a√±os\n";
    echo "  Notas: " . implode(", ", $estudiante["notas"]) . "\n";
    
    $promedio = array_sum($estudiante["notas"]) / count($estudiante["notas"]);
    echo "  Promedio: " . number_format($promedio, 2) . "\n\n";
}
?>
'''

resultado = ejecutar_php(codigo)
print("üì§ Salida:")
print(resultado)
```

## 4. ‚öôÔ∏è Funciones

Las funciones permiten reutilizar c√≥digo y organizarlo mejor.

### Caracter√≠sticas:
- **Par√°metros opcionales**: Con valores por defecto
- **Tipado de par√°metros**: Desde PHP 7+
- **Funciones an√≥nimas**: Closures
- **Funciones variables**: Llamar funciones din√°micamente

```python
codigo = '''
<?php
function saludar($nombre, $apellido = "An√≥nimo") {
    return "Hola, $nombre $apellido!";
}

function calcularAreaCirculo($radio) {
    return pi() * pow($radio, 2);
}

function esPar($numero) {
    return $numero % 2 == 0;
}

echo "=== FUNCIONES B√ÅSICAS ===\n";
echo saludar("Ana") . "\n";
echo saludar("Carlos", "Garc√≠a") . "\n";

echo "\n=== C√ÅLCULOS ===\n";
$radio = 5;
echo "√Årea de c√≠rculo (radio $radio): " . number_format(calcularAreaCirculo($radio), 2) . "\n";

echo "\n=== VALIDACIONES ===\n";
for ($i = 1; $i <= 5; $i++) {
    echo "$i es " . (esPar($i) ? "par" : "impar") . "\n";
}
?>
'''

resultado = ejecutar_php(codigo)
print("üì§ Salida:")
print(resultado)
```

## 5. üèóÔ∏è Programaci√≥n Orientada a Objetos

PHP soporta completamente la POO con clases, herencia, interfaces, traits, etc.

### Conceptos clave:
- **Clases y objetos**
- **Herencia**
- **Encapsulamiento**
- **Polimorfismo**
- **Interfaces y traits**

```python
codigo = '''
<?php
class Vehiculo {
    protected $marca;
    protected $modelo;
    protected $a√±o;
    
    public function __construct($marca, $modelo, $a√±o) {
        $this->marca = $marca;
        $this->modelo = $modelo;
        $this->a√±o = $a√±o;
    }
    
    public function obtenerInfo() {
        return $this->marca . " " . $this->modelo . " (" . $this->a√±o . ")";
    }
    
    public function acelerar() {
        return "El veh√≠culo est√° acelerando";
    }
}

class Coche extends Vehiculo {
    private $puertas;
    
    public function __construct($marca, $modelo, $a√±o, $puertas = 4) {
        parent::__construct($marca, $modelo, $a√±o);
        $this->puertas = $puertas;
    }
    
    public function obtenerInfo() {
        return parent::obtenerInfo() . " - {$this->puertas} puertas";
    }
    
    public function tocarClaxon() {
        return "¬°Beep beep!";
    }
}

echo "=== PROGRAMACI√ìN ORIENTADA A OBJETOS ===\n";
$miCoche = new Coche("Toyota", "Corolla", 2023, 4);
echo "Mi coche: " . $miCoche->obtenerInfo() . "\n";
echo $miCoche->acelerar() . "\n";
echo $miCoche->tocarClaxon() . "\n";

$deportivo = new Coche("Ferrari", "F8", 2024, 2);
echo "\nDeportivo: " . $deportivo->obtenerInfo() . "\n";
?>
'''

resultado = ejecutar_php(codigo)
print("üì§ Salida:")
print(resultado)
```
